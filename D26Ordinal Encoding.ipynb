```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('customer.csv')
```

```python
df=df.iloc[:,2:]
```

```python
df.head(33)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2) ##inside train_test_split(x,y) x-->kota row kota col thakbe(parameters)   y-->kota row r col thakbe(features) -1=last column 
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])  ##'School'<'UG'<'PG'
```

```python
oe.fit(X_train)
```

```python
X_train=oe.transform(X_train)
X_test=oe.transform(X_test)
```
